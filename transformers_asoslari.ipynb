```python
from transformers import pipeline
```

```python
# Santiment (Hissiyotni aniqlash)

classifier = pipeline("sentiment-analysis")

result1 = classifier("I've been waiting for a HuggingFace course my whole life.")

result2 = classifier("This site is difficult to use as some parts are working")

print(result1)
print(result2)
```

```python
# Fill-mask (Niqoblangan so'zni to'ldirish)

fill_mask = pipeline('fill-mask')

result1 = fill_mask('Deep Learning is <mask>')

result2 = fill_mask('The capital of Uzbekistan <mask>')

print(result1)
print(result2)
```

```python
# Named Entity Recognition (NER Nomli obyeklarni aniqlash)

ner = pipeline('ner')

result1 = ner('My name is Saidakbar. I\'m university is Alfraganus')

print(result1)
```
